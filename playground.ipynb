```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import json
from IPython.display import display, HTML
```

```python
selected = 'ETH'
```

```python
fp = open(f'app/_layer/simresults/{selected}_points.json',)
points = json.load(fp)
fb = open(f'app/_layer/simresults/{selected}_buys.json',)
buys = json.load(fb)
fs = open(f'app/_layer/simresults/{selected}_sells.json',)
sells = json.load(fs)

points = pd.DataFrame(points)
buys = pd.DataFrame(buys)
sells = pd.DataFrame(sells)
```

```python
plt.figure(figsize=(16, 8))
plt.plot(points[[0]], points[[1]], c='lightblue') 
plt.scatter(buys[[0]], buys[[1]], c='green')
plt.scatter(sells[[0]], sells[[1]], c='red')
plt.show()
```
